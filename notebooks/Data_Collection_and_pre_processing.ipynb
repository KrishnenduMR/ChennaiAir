```python
import os
import pandas as pd

# Define data directory
data_dir = "../data/CPCBData/"

# List of stations and years
stations = ["alandur", "velachery", "manali"]
years = ["2021", "2022", "2023", "2024"]

# Load and combine data
dataframes = []
for station in stations:
    for year in years:
        file_path = os.path.join(data_dir, f"{station}_{year}.csv")
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            df["station"] = station  # Add station column
            dataframes.append(df)

# Combine all DataFrames
df_combined = pd.concat(dataframes, ignore_index=True)
```

```python
# Convert timestamp
df_combined["Timestamp"] = pd.to_datetime(df_combined["Timestamp"])

# Sort by time
df_combined = df_combined.sort_values(by=["station", "Timestamp"])

# Forward and backward fill
df_combined.ffill(inplace=True)
df_combined.bfill(inplace=True)


# Fill remaining missing values with station-wise mean
df_combined.fillna(df_combined.groupby("station").transform("mean"), inplace=True)
```

```python
df_combined
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check missing values
print(df_combined.isnull().sum())

# Line plot for PM2.5
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_combined, x="Timestamp", y="PM2.5 (µg/m³)", hue="station")
plt.title("PM2.5 Concentration Over Time")
plt.show()

# Distribution plot for NO2 per station
plt.figure(figsize=(12, 6))
sns.boxplot(x="station", y="NO2 (µg/m³)", data=df_combined)
plt.title("NO2 Concentration by Station")
plt.show()
```

```python
# Time-based features
df_combined["hour"] = df_combined["Timestamp"].dt.hour
df_combined["day"] = df_combined["Timestamp"].dt.day
df_combined["month"] = df_combined["Timestamp"].dt.month
df_combined["year"] = df_combined["Timestamp"].dt.year

# Wind speed interaction
df_combined["Wind Impact"] = df_combined["WS (m/s)"] * df_combined["WD (deg)"]

# Normalize pollutant concentrations
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
pollutants = ["PM2.5 (µg/m³)", "PM10 (µg/m³)", "NO2 (µg/m³)", "SO2 (µg/m³)", "CO (mg/m³)", "Ozone (µg/m³)"]
df_combined[pollutants] = scaler.fit_transform(df_combined[pollutants])
```

```python
df_combined
```

```python
# Remove outliers using IQR method
Q1 = df_combined[pollutants].quantile(0.25)
Q3 = df_combined[pollutants].quantile(0.75)
IQR = Q3 - Q1

df_combined = df_combined[~((df_combined[pollutants] < (Q1 - 1.5 * IQR)) | 
                            (df_combined[pollutants] > (Q3 + 1.5 * IQR))).any(axis=1)]
df_combined
```

```python
# Calculate missing value percentage
missing_percentage = (df_combined.isnull().sum() / len(df_combined)) * 100

# Display columns with missing values
missing_percentage = missing_percentage[missing_percentage > 0]
print(missing_percentage.sort_values(ascending=False))
```

```python
# List of columns to drop
cols_to_drop = ["Xylene (µg/m³)", "AT (°C)", "RF (mm)", "TOT-RF (mm)"]

# Drop columns
df_combined.drop(columns=cols_to_drop, inplace=True)

# Verify
print("Remaining columns:", df_combined.columns)
```

```python
# Convert NOx from ppb to µg/m³
df_combined["NOx (µg/m³)"] = df_combined["NOx (ppb)"] * 1.88

# Convert CO from mg/m³ to µg/m³
df_combined["CO (µg/m³)"] = df_combined["CO (mg/m³)"] * 1000

# Convert BP from mmHg to hPa
df_combined["BP (hPa)"] = df_combined["BP (mmHg)"] * 1.33322

# Drop original columns if necessary
df_combined.drop(columns=["NOx (ppb)", "CO (mg/m³)", "BP (mmHg)"], inplace=True)
```

```python
df_combined.to_csv("../data/CombinedData/cleaned_air_quality.csv", index=False)
```
