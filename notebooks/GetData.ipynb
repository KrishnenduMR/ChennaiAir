```python
df_man = pd.read_csv("../data/RAW_DATA/manali_data.csv")
df_man
```

```python
df_vela = df_vela.loc[:, ~df_vela.columns.duplicated()]
df_ala = df_ala.loc[:, ~df_ala.columns.duplicated()]
df_man = df_man.loc[:, ~df_man.columns.duplicated()]
```

```python
df_ala['Station'] = 'Alandur'
df_vela['Station'] = 'Velachery'
df_man['Station'] = 'Manali'
```

```python
# For df_vela and df_ala
df_vela.columns.difference(df_ala.columns)

# For df_ala and df_vela
df_ala.columns.difference(df_vela.columns)

# For df_vela and df_man
df_vela.columns.difference(df_man.columns)

# For df_man and df_vela
df_man.columns.difference(df_vela.columns)

# For df_ala and df_man
df_ala.columns.difference(df_man.columns)

# For df_man and df_ala
df_man.columns.difference(df_ala.columns)
```

```python
total_rows = df_vela.shape[0] + df_ala.shape[0] + df_man.shape[0]
total_rows
```

```python
frames = [
    df_vela,
    df_ala,
    df_man
]

df_combined = pd.concat(frames, axis=0)
df_combined
```

```python
df_combined.columns
```

```python
df_combined = df_combined.loc[:, ~df_combined.columns.isna()]
```

```python
df_combined.columns # NaN is removed
```

```python
required_columns = [
    'From Date', 'To Date', 'Station', 'PM2.5', 'NO', 'NO2', 'NOx', 'SO2', 'CO',
    'Ozone', 'Benzene', 'Toluene', 'Eth-Benzene', 'MP-Xylene', 'O', 'Temp', 'RH',
    'WS', 'WD', 'SR', 'BP', 'VWS', 'Xylene', 'AT', 'RF', 'TOT-RF', 'PM10', 'NH3', 'CO2'
]

# Keep only the columns that exist in the dataframe
existing_columns = [col for col in required_columns if col in df_combined.columns]
df_combined = df_combined[existing_columns]
```

```python
df_combined = df_combined.reset_index()
df_combined.drop(['index'], axis = 1, inplace = True)
df_combined
```
