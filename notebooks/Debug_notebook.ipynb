```python
import os
import pandas as pd

# Define AQI data directory
aqi_data_dir = "../data/AQIData/"

# List of stations
stations = ["alandur", "velachery", "manali"]

# Process each station separately
for station in stations:
    hourly_path = os.path.join(aqi_data_dir, f"{station}_hourly_aqi.csv")
    daily_path = os.path.join(aqi_data_dir, f"{station}_daily_aqi.csv")

    if os.path.exists(hourly_path):
        df_hourly = pd.read_csv(hourly_path)
        print(f"\nHourly AQI Data for {station}:")
        print(df_hourly.head())
        print(df_hourly.info())

    if os.path.exists(daily_path):
        df_daily = pd.read_csv(daily_path)
        print(f"\nDaily AQI Data for {station}:")
        print(df_daily.head())
        print(df_daily.info())
```
